# Plots Through Time

```
import plotly.graph_objects as go
import plotly.express as px
```

## Houly

```
drugs = df_houly.columns[1:9]
```

```
# fig = go.Figure()
# for drug in drugs:
#   fig.add_trace(go.Scatter(x= list(range(len(df_houly))),
#                            y = df_houly.loc[:,drug],
#                            mode = 'lines+markers',
#                            name = drug))
# fig.show()
```

## Daily

```
fig = go.Figure()
for drug in drugs:
  fig.add_trace(go.Scatter(x= list(range(len(df_daily))),
                           y = df_daily.loc[:,drug],
                           mode = 'lines+markers',
                           name = drug))
fig.show()
```

## Weekly

```
fig = go.Figure()
for drug in drugs:
  fig.add_trace(go.Scatter(x= list(range(len(df_weekly))),
                           y = df_weekly.loc[:,drug],
                           mode = 'lines+markers',
                           name = drug))
fig.show()
```

## Monthly

```
fig = go.Figure()
for drug in drugs:
  fig.add_trace(go.Scatter(x= list(range(len(df_monthly))),
                           y = df_monthly.loc[:,drug],
                           mode = 'lines+markers',
                           name = drug))
fig.show()
```

# Quantity Plots

```
fig = px.bar(df_monthly,x=drugs,y =  list(df_monthly.iloc[:,1:].sum()))
fig.show()
```
